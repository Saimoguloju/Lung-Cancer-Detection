```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['GENDER'] = label_encoder.fit_transform(df['GENDER'])
df['SMOKING'] = label_encoder.fit_transform(df['SMOKING'])
df['YELLOW_FINGERS'] = label_encoder.fit_transform(df['YELLOW_FINGERS'])
df['ANXIETY'] = label_encoder.fit_transform(df['ANXIETY'])
df['PEER_PRESSURE'] = label_encoder.fit_transform(df['PEER_PRESSURE'])
df['CHRONIC_DISEASE'] = label_encoder.fit_transform(df['CHRONIC_DISEASE'])
df['FATIGUE'] = label_encoder.fit_transform(df['FATIGUE'])
df['ALLERGY'] = label_encoder.fit_transform(df['ALLERGY'])
df['WHEEZING'] = label_encoder.fit_transform(df['WHEEZING'])
df['ALCOHOL_CONSUMING'] = label_encoder.fit_transform(df['ALCOHOL_CONSUMING'])
df['COUGHING'] = label_encoder.fit_transform(df['COUGHING'])
df['SHORTNESS_OF_BREATH'] = label_encoder.fit_transform(df['SHORTNESS_OF_BREATH'])
df['SWALLOWING_DIFFICULTY'] = label_encoder.fit_transform(df['SWALLOWING_DIFFICULTY'])
df['CHEST_PAIN'] = label_encoder.fit_transform(df['CHEST_PAIN'])
df['LUNG_CANCER'] = label_encoder.fit_transform(df['LUNG_CANCER'])
```

```python
for col in df:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.savefig("bar_graph.png")
    plt.show()
```

```python
x = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,shuffle = True, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
pred_lr = LR.predict(x_test)
pred_lr
```

```python
lr_acc = LR.score(x_test,y_test)
lr_acc
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
pred_dt = DT.predict(x_test)
pred_dt
```

```python
dt_acc = DT.score(x_test,y_test)
dt_acc
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(x_train, y_train)
```

```python
predict_gbc = GBC.predict(x_test)
predict_gbc
```

```python
gbc_acc = GBC.score(x_test, y_test)
gbc_acc
```

```python
print(classification_report(y_test,predict_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state=0)
RF.fit(x_train,y_train)
```

```python
predict_rf = RF.predict(x_test)
predict_rf
```

```python
rf_acc = RF.score(x_test, y_test)
rf_acc
```

```python
print(classification_report(y_test,predict_rf))
```

```python
models = pd.DataFrame({
    'Model' : ['Logistic Regression','Decision Tree Classifier', 'GradientBoostingClassifier','Random Forest Classifier'],
    'Accuracy Score' : [lr_acc, dt_acc, gbc_acc, rf_acc]})
     
    
models.sort_values(by = 'Accuracy Score', ascending = False)
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=models.sort_values(by = 'Accuracy Score', ascending = False), x='Model', y='Accuracy Score')
plt.title('Accuracy Scores of Different Models')
plt.ylabel('Accuracy Score')
plt.savefig("Accuracy Scores of Different Models.png")
plt.show()
```

## Conclusion
1. GradientBoostingClassifier: This model achieved the highest accuracy score of 0.518333, indicating it performed the best on your dataset compared to the other models.

2. Logistic Regression: With an accuracy of 0.506667, this model also performed relatively well but not as well as Gradient Boosting.

3. Decision Tree Classifier: This model had an accuracy of 0.501667, which is slightly above random guessing (assuming a binary classification problem).

4. Random Forest Classifier: This model had the lowest accuracy score of 0.500000, which suggests it performed no better than random guessing for your dataset.
